# CSDA Satellogic Evalution Image Data

```r
library(ggplot2)
library(dplyr)
library(RColorBrewer)
```

```r
# Read the CSV file
df <- read.csv("/projects/my-private-bucket/databank/csda/NASA CSDA Multi-Spectral Sensor Platform Metadata - Evaluation Image Table.csv")
```

```r
options(repr.plot.width = 8, repr.plot.height = 4, repr.plot.res = 300)

set2_new <- colorRampPalette(brewer.pal(8, "Set1"))(length(levels(factor(df$Sensor))))

# Create the bar plot
ggplot(df, aes(y = Site.Name, fill = Sensor)) +
  geom_bar() +
  scale_fill_manual(values=set2_new) +
  facet_wrap(Evaluation.Type~ Generation, nrow=1) +
  labs(
    title = paste0("Evaluating a constellation: ",levels(factor(df$Parent.Company))),
    y = "Evaluation Site Name",
    x = "Number of Images",
    #fill = "Generation"
  ) +
  theme_bw() +
  theme(
    axis.text.x = element_text(angle = 45, hjust = 1),
    plot.title = element_text(hjust = 0.5),
    strip.text = element_text(face = "bold"),
    legend.position = "top",
    legend.direction = "horizontal"
  ) +
  guides(fill = guide_legend(ncol = 8))
```
